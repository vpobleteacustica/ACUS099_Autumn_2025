### 1) Propiedad de traslado (shifting) en el Tiempo
> <font size="4"> + Trasladar en el tiempo es lo mismo que multiplicar por una exponencial compleja de signo negativo en el dominio de frecuencia:</font>

$$
\begin{align*}
		\large x[n-n_0] & \large \overset{\mathscr{ \large  F}}{\longleftrightarrow}\large e^{-j\omega n_0}\cdot X(e^{j\omega}) 
\end{align*}
$$

> + <font size="4">**Demostración:**</font>
$$
\begin{align*}
\mathscr{ \large  F}\Big\{\large x[n-n_0]\Big\} & \large{= \sum\limits_{n=-\infty}^{\infty} x[n-n_0]e^{-j\omega n}} \\
& \large{=  \sum\limits_{m=-\infty}^{\infty} x[m]e^{-j\omega \,(m+n_0)}} \quad ({\text{donde sustituimos}}\,\, m=n-n_0)\\
&  {\large = e^{-j\omega n_0}} \large{\sum\limits_{m=-\infty}^{\infty} x[m]e^{-j\omega m}} \\
&  {\large = e^{-j\omega n_0}} \cdot{\large X(e^{j\omega})}
\end{align*}
$$

### 2) Propiedad de Traslado (shifting) en la Frecuencia

> <font size="4"> + Trasladar en la frecuencia es lo mismo que multiplicar por una exponencial compleja de signo positivo en el dominio del tiempo:</font>

$$
\begin{align*}
		\large x[n]\cdot e^{j\omega_0 n} & \large \overset{\mathscr{ \large  F}}{\longleftrightarrow}\large X(e^{j(\omega-\omega_0)}) 
\end{align*}
$$

> + <font size="4">**Demostración:**</font>
$$
\begin{align*}
\mathscr{ \large  F}\Big\{\large x[n]\cdot e^{j\omega_0 n}\Big\} & \large{=\sum\limits_{n=-\infty}^{\infty} x[n]\cdot e^{j\omega_0 n}\cdot e^{-j\omega n}} \\
&  \large {= \sum\limits_{n=-\infty}^{\infty} x[n]e^{-j(\omega-\omega_0) n}} \\
& {\large \equiv } \,{\large X(e^{j(\omega-\omega_0)})}
\end{align*}
$$


### 3) Modelado del proceso de muestreo con un tren de impulsos periódicos: $s(t)=\sum_{n=-\infty}^{\infty} \delta(t-nT)$

> + Observamos en la Figura que $s(t)$ es periódica en el tiempo y su período es $T$.

> + El **Teorema del muestreo de Nyquist establece que** para reproducir adecuadamente una señal originalmente continua, ésta debe ser muestreada en forma periódica a una tasa 2 veces la frecuencia más alta que uno desea grabar.

### 4) Multiplicación en el tiempo, convolución en el dominio de la frecuencia

> + La convolución de dos secuencias puede ser calculada ya sea: 1) en dominio del tiempo; o bien, 2) en el dominio de la frecuencia. Si recordamos el teorema de convolución, éste nos permite calcular la convolución de dos secuencias simplemente multiplicando sus espectros.

> + Existe un teorema adicional que involucra a la convolución en el dominio de la frecuencia. Este teorema establece que la multiplicación de dos secuencias en el tiempo es equivalente en el dominio de la frecuencia a la convolución de sus espectros.

> + Matemáticamente, si:
$$
\begin{align*}
		x[n] & \overset{\mathscr{F}}{\longleftrightarrow} X(e^{j\omega}) 
\end{align*}
$$

y
$$
\begin{align*}
		{\text{w}}[n] & \overset{\mathscr{F}}{\longleftrightarrow} W(e^{j\omega}) 
\end{align*}
$$

> + entonces

$$
    \begin{align}
		x[n]\,{\text{w}}[n] & \overset{\mathscr{F}}{\longleftrightarrow} \frac{1}{2\pi}X(e^{j\omega})*W(e^{j\omega}) = \frac{1}{2\pi}\int_{-\pi}^{\pi} X(e^{j\theta})\, W(e^{j(\omega - \theta})\,d\theta     
	\label{tcf}
	\end{align}
$$

> + La Ec. (2) es una convolución de las correspondientes transformadas de Fourier, ambas funciones periódicas en frecuencia, de período $2\pi$. Los límites de integración en la convolución se extienden sólo sobre un período: $-\pi \leq \omega\leq \pi$.

> + En efecto: 
$$
\begin{align*}
		\mathscr{F}\left\{x[n]{\text{w}}[n]\right\}	& = \sum_{n=-\infty}^{\infty} x[n]{\text{w}}[n]\,e^{-j\omega n}\\
		&= \sum_{n=-\infty}^{\infty}\Big( \frac{1}{2\pi}\int_{-\pi}^{\pi}X(e^{j\theta})\,e^{j\theta n}d\theta\Big)\,{\text{w}}[n]\,e^{-j\omega n}\\
		& = \frac{1}{2\pi}\int_{-\pi}^{\pi}X(e^{j\theta})\, \sum_{n=-\infty}^{\infty}{\text{w}}[n]\,e^{-j(\omega - \theta) n}\,d\theta\\
		&= \frac{1}{2\pi}\int_{-\pi}^{\pi}X(e^{j\theta})\,W(e^{j(\omega - \theta)})\,d\theta \\
		&=\frac{1}{2\pi}\Big[X(e^{j\omega})*W(e^{j\omega})\Big]
	\end{align*}
$$

> + Por lo tanto,

$$
\begin{align}
		x[n]\,{\text{w}}[n] & \overset{\mathscr{F}}{\longleftrightarrow} \frac{1}{2\pi}\Big[X(e^{j\omega})*W(e^{j\omega})\Big]     
	\end{align}
$$

> + Ahora, como estable el teorema del muestreo de Nyquist: si $x_c(t)$ es la señal de banda limitada al intervalo de frecuencias: $(-\Omega_N,\Omega_N)$, donde $\Omega_N$ es la frecuencia máxima a grabar, entonces esta señal continua se podrá reconstruir a partir de las muestras $x(nT)$, siempre y cuando la frecuencia de muestreo $\Omega_s = 2\pi /T$, satisfaga la condición:

$$
    \begin{align}
        \Omega_s & = \frac{2\pi}{T} > 2\Omega_N 
    \end{align}
    $$
> + O bien,
$$     
     \begin{align}
        (\Omega_s -\Omega_N) & > \Omega_N 
    \end{align}
$$

> + Se llama habitualmente a $\Omega_N$ como la frecuencia frecuencia de Nyquist.

### 5) Modelado del proceso de muestreo:

> + Este proceso de muestreo se puede modelar usando un tren de impulsos $s(t)$ en el tiempo, que multiplica a la función continua $x_c(t)$ entonces:
    
\begin{align*}
x_s(t)&= x_c(t)\cdot s(t)        
\end{align*}

> + Anaĺicemos el tren de impulsos en el tiempo:

\begin{align*}
        s(t) &= \sum_{n=-\infty}^{\infty} \delta(t-nT)        
    \end{align*}

> + Dado que $s(t)$ es periódica en el tiempo, de período $T$, podemos expresarla por su serie de Fourier:

\begin{align*}
            s(t) &= \sum_{k=-\infty}^{\infty} a_k\, e^{jk\Omega_s t}
\end{align*}

> + donde los coeficientes $a_k$ se pueden encontrar integrando $s(t)$ sobre un período:

\begin{align*}
            a_k &= \frac{1}{T}\int_{-\frac{T}{2}}^{\frac{T}{2}} \delta(t)\,e^{-jk\Omega_s t}\,dt\\
            &= \frac{1}{T}
        \end{align*}

> + valor que es independiente de $k$ y por lo tanto es el mismo para todo coeficiente $a_k$.
    
> + Así que:
    
\begin{align*}
            s(t) &= \frac{1}{T} \sum_{k=-\infty}^{\infty}\, e^{jk\Omega_s t}
\end{align*}
    
> + Aplicamos $\mathscr{F}\left\{\cdot\right\}$ a ambos lados para obtener:

\begin{align*}
\mathscr{ \large  F}\left\{s(t)\right\} &= \frac{1}{T} \sum_{k=-\infty}^{\infty}\,\mathscr{ \large  F}\left\{ e^{jk\Omega_s t}\right\}\\
S(j\Omega) &=  \frac{1}{T} \sum_{k=-\infty}^{\infty}\, 2\pi\,\delta(\Omega - k\Omega_s)
\end{align*}

> + donde $\Omega_s = 2\pi/T$, luego

\begin{align*}
           S(j\Omega) &=  \frac{2\pi}{T} \sum_{k=-\infty}^{\infty}\, \delta(\Omega - k\Omega_s)
\end{align*}

> + Con este resultado, vemos que en el dominio de la frecuencia, el modelo usado para representar el proceso de muestreo será:

\begin{align*}
           \large x_s(t)&= \large x_c(t)\cdot s(t) \,\,\,\,\,\,\large/\mathscr{ \large  F}\left\{\cdot\right\}\\
           \mathscr{ \large  F}\large \left\{x_s(t)\right\} &= \mathscr{ \large  F}\large \left\{x_c(t)\cdot s(t)\right\}\\
           X_s(j\Omega) &= \frac{1}{2\pi}\Big[X_c(j\Omega)*S(j\Omega)\Big]\\
           &=\frac{1}{2\pi}\Big[X_c(j\Omega)*\frac{2\pi}{T} \sum_{k=-\infty}^{\infty}\, \delta(\Omega - k\Omega_s)\Big]\\
           &=\frac{1}{T}\,\sum_{k=-\infty}^{\infty}\Big[X_c(j\Omega)*\delta(\Omega - k\Omega_s)\Big]
\end{align*}

> + donde podemos usar la convolución en el dominio de tiempo continuo:

\begin{align*}
		\int_{-\infty}^{\infty}\, X_c(\eta)\,\delta(\Omega - k\Omega_s - \eta)\,d\eta & = X_c(\Omega - k\Omega_s) 
\end{align*}

> + Por lo tanto,
	
\begin{align*}
		X_s(j\Omega)&= \frac{1}{T}\,\sum_{k=-\infty}^{\infty}X_c(\Omega - k\Omega_s)  
\end{align*}

> + La relación entre la frecuencia en tiempo continuo, $\Omega$, y la frecuencia en tiempo discreto, $\omega$, es $\omega=\Omega T$. Por lo tanto, vemos que,

\begin{align*}
        X_s(j\Omega) &= X(e^{j\omega})\,|_{\omega=\Omega T}
\end{align*}

> + Consecuentemente,
    \begin{align*}
        X(e^{j\omega}) &=\frac{1}{T}\,\sum_{k=-\infty}^{\infty}X_c(j\frac{\omega}{T} - j\frac{2\pi k}{T})
    \end{align*}

> + Interpretación: 
>> + El espectro de la salida del modelo consiste de copias repetidas del espectro de la señal de entrada al modelo.
>> + Estas copias están shifteadas por múltiplos enteros de la frecuencia de muestreo  y luego son sumadas esas copias.
